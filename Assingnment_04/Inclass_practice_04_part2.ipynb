```python
def cleanse_year(year_str):
    if "*" in year_str:
        year = year_str.replace("*", "")
    else:
        year = year_str

    return int(year)
```

```python
print(cleanse_year("2010*"))
print(type(cleanse_year("2010*")))
```

```python
print(cleanse_year("2000"))
print(type(cleanse_year("2000")))
```

```python
df["year_cleansed"] = df["Census Year"].apply(cleanse_year)

df
```
