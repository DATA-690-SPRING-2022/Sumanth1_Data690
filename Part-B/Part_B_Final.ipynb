```python
import pandas as pd
import plotly
import plotly.express as px
DATA_URL = "https://raw.githubusercontent.com/DATA-690-SPRING-2022/Sumanth1_Data690/main/Part-B/wdi_data.csv"
DATA_URL = "https://raw.githubusercontent.com/DATA-690-SPRING-2022/Sumanth1_Data690/main/Part-B/wdi_Urban%20data.csv"
DATA_URL = "https://raw.githubusercontent.com/DATA-690-SPRING-2022/Sumanth1_Data690/main/Part-B/wdi_Rural%20data.csv"
```

```python
df = pd.read_csv(DATA_URL)

df.sample(10)
```

```python
df_2016_electricity_all = df.query("Year == 2016").query("indicator == 'EG.ELC.ACCS.ZS'")
df_2016_electricity_all.sample(5)
```

```python
df_2016_electricity_all_sort = df_2016_electricity_all.sort_values(by="value", ascending=False)

df_2016_electricity_all_sort.head()
```

```python
fig = px.bar(
    df_2016_electricity_all_sort, 
    x="Country Name", 
    labels={"value":"2016 Bar Plot of all countries electricity percentage", "Country Name":""},
    y="value",
    color="Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5]
)

fig.update_layout(showlegend=False)

fig.show()
```

# Access to electricity, urban(% of Urban population)

```python
df_2016_Urban_all = df.query("indicator == 'EG.ELC.ACCS.UR.ZS'")
df_2016_Urban_all = df_2016_Urban_all[df_2016_Urban_all["Country Code"].isin(["USA","CAN", "LKA", "PAK","AFG"])]

fig = px.line(
    df_2016_Urban_all, 
    x="Year", 
    labels={"value":"electricity % of urban population"},
    y="value",
    color="Country Name",
    height=800,
    template=list(plotly.io.templates.keys())[5],
    title="TOP 5 Countries by 2016 Access to electricity (% of population)"

)


fig.update_layout(showlegend=True)

fig.show()
```

* I used a Line Graph to show the difference in urban population using eletricity between Canada, United States, Afghanistan, Pakistan and Sri Lanka over time.
* People in USA and Canada were using the 100% of the electricity. USA and Canada were utilizing the electricity in all aspects.
* There was very much fluctuation in Afghanistan and from the year 2011 to 2018 there are severe up and downs in the usage of energy.
* Urban population in Pakistan and Sri Lanka were slightly increasing in terms of eletrical usage.
